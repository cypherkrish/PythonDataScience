# Sub-Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = x * 2
```

```python
fig, axes = plt.subplots(nrows = 1 , ncols =2 )
axes[0].plot(x, y)
axes[1].plot(a, b)
plt.show()
```

```python
plt.show()
plt.show()
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(a, b)
axes[0][0].set_title("Title 1 for graph 1")

axes[1][0].plot(a, b)
axes[1][0].set_ylabel("Y label is b")
axes[1][0].set_xlabel("X label is a")
axes[1][0].set_xlim(3, 7)


axes[0][1].plot(a, b)
axes[1][1].plot(x, y)

fig.subplots_adjust(wspace = 0.5, hspace = 0.5 )

fig.suptitle("Figure level TITLE :)", fontsize = 20)
#plt.tight_layout()
plt.show()


fig.savefig("newsubplot.png", bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

axes[0].plot(a, b)
axes[1].plot(x, y)
axes[2].plot(a, b)

plt.tight_layout()
plt.show()
```

# Visual styling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X', color = '#4bab3f', lw = 2, ls = '--')   # Pass the hex code for the color

ax.plot(x, x+1, label = 'X vs X + 1', color = '#ab3f60', lw = 10, linestyle = '-.')

ax.plot(x, x * 2, label = 'X vs X + 1', color = '#f4dd0a', linewidth = 5, linestyle = ':')

ax.plot(x, x+3, label = 'X vs X + 3', color = '#f9043b', lw = 2, marker = '+', markersize = 20)[0].set_dashes([10, 10, 5, 1])


ax.plot(x+1, x * 2, label = 'X vs X + 3', color = 'purple', lw = 2, marker = 's', ls = '--', markersize = 20, markerfacecolor = 'red', markeredgewidth = 5, markeredgecolor = 'orange')

ax.legend()
```

```python
# Example find E = m * (c ** 2 ) for 10 diffeent massses from 0 to 10

import numpy as np
import matplotlib.pyplot as plt
```

```python
m  = np.linspace(0, 10, 11)
```

```python
m
```

```python
c = 3 * (10 ** 8)
```

```python
E = m * c ** 2
```

```python
E
```

```python
# plotting the graph

# Envergy (joules) in y axis and mass on x-axis
```

```python
plt.plot(m, E, color = 'red', linewidth = 5)
plt.title('E = m * c ^ 2')
plt.xlabel ('M in g')
plt.ylabel ('E in jules')

plt.xlim(0, 10)
plt.show()
```

```python
# The same plot cab be created using fig function

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(m, E, color = 'red', lw = 5)
ax.set_xlabel("Mass in grms")
ax.set_ylabel("Energy in joules")
```

```python
# plot in the log scale
```

```python
plt.plot(m, E, color = "red", lw = 5)
plt.title("E = m * c ^ 2")
plt.xlabel("Mass in grams")
plt.ylabel("Engergy in Jules")
plt.xlim(0, 10)

plt.yscale("log")
plt.grid(which = 'both',axis = 'y' )
```

```python
labels = ['1 mo', '2 Mo', '3 Mo', '4 Mo', '5 Mo' ]
year1 = [2.57, 8.98, 7.85, 9.56, 10.26]
year2 = [0.2, 0.25, 1.25, 0.15, .098]
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1 , 1])

ax.plot(labels, year1, label = 'Year 1')
ax.plot(labels, year2, label = 'Year 2')
ax.legend(loc =(1.2, 0.5))
plt.show()
```

```python
# using the subplots
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 5))

axes[0].plot(labels, year1)
axes[0].set_title("Year 1")
axes[1].plot(labels, year2)
axes[1].set_title("Year 2")

#ax.legend(loc =(1.2, 0.5))
plt.tight_layout()
```
